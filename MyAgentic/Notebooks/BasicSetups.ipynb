### Text Loading...

```python
from langchain_community.document_loaders import TextLoader
pathtxt = os.path.join("..","..","Data","ClimateChange.txt")
```

```python
# Normalize and convert utf-8
import chardet
with open(pathtxt,"rb") as corsur:
    raw = corsur.read(1000)
enc = chardet.detect(raw)['encoding'] or "utf-8"
```

```python

text_loader  = TextLoader(pathtxt,encoding=enc)
txtdoc       = text_loader.load()
```

```python
txtdoc[0].page_content
```
